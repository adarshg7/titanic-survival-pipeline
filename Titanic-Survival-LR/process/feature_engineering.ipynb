```python
import pandas as pd
df_train = pd.read_csv('../DataSet/raw data/train.csv')
df_test = pd.read_csv('../DataSet/raw data/test.csv')
df = pd.concat([df_train,df_test], axis = 0)
```

```python
safe_df = df
```

```python
df.info()
```

```python
df.drop(['PassengerId','Name'],axis =1, inplace=True)
```

```python
df.info()
```

```python
df['Sex'].unique()
```

```python
df['Sex'] = df['Sex'].map({'female':0, 'male':1})
```

```python
df.info()
```

```python
df['Deck'] = df['Cabin'].str[0].fillna('U')
```

```python
df.info()
```

```python
df['Deck_Numeric'] = df['Deck'].map({
    'B': 7, 'C': 6, 'D': 5, 'E': 4, 'A': 3, 'F': 2, 'G': 1, 'U': 0
})
```

```python
df.drop('Deck',inplace=True,axis=1)
```

```python
df.info()
```

```python
df['Cabin'].unique()
```

```python
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
df.drop(['Cabin'], inplace=True, axis=1)
```

```python
df.info()
```

```python
df.drop('Ticket', inplace=True, axis=1)
```

```python
df.info()
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'] = df['Embarked'].map({'C':0,'Q':1,'S':2})
```

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Deck_Numeric'] = df['Deck_Numeric'].fillna(0)
```

```python
df.info()
```

```python
df.to_csv("../DataSet/cleaned_data/train_clean.csv", index=False)
```

```python
train_df = df.iloc[:891].copy()
test_df = df.iloc[891:].copy()
test_df = test_df.drop(['Survived'], axis =1)
```

```python
# Save the cleaned training data
train_df.to_csv('../DataSet/cleaned_data/train_clean.csv', index=False)

# Save the cleaned test data
test_df.to_csv('../DataSet/cleaned_data/test_clean.csv', index=False)
```
